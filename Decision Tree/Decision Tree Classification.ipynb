```python
import pandas as pd
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset.columns
```

```python
dataset=dataset.drop(columns=["User ID"])
```

```python
dataset.columns
```

```python
independent = dataset[['Age', 'EstimatedSalary','Gender_Male']]
dependent = dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=1/3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='gini',splitter='best',random_state=0)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
clf_report = classification_report(y_test, y_pred)
```

```python
cm
```

```python
print(clf_report)
```

precision class 0 = 78/(78+11) = 0.88
precision class 1 = 38/(38+7) = 0.84

recall class 0 = 78/(78+7) = 0.92
recall class 1 = 38/(38+11) = 0.78

F1 score class 0 = 2*(precision*recall) / (precision+recall) = 2*(0.88*0.92)/(0.88+0.92) =0.90
F1 score class 1 = 2*(0.84*0.78)/(0.84+0.78) = 0.81

Accuracy = (True Positive+True Negative) / All = (78+38)/(78+38+11+7) = 0.87

Recall :(Missed cases) How much percentage of correctly predicted out of Actual value in each section? How much percentage of users can buy the product from the Ad?

Precision:(False Alarm) How much percentage of correctly predicted out of total prediction in each section? what percentage of potential buyers missed to display the Ad?
