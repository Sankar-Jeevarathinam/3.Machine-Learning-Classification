```python
import pandas as pd
dataset = pd.read_csv('CKD.csv')
```

```python
dataset.shape
```

```python
dataset.head(5)
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
dependent.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
param_grid={'criterion':['gini', 'entropy', 'log_loss'],
           'splitter':['best', 'random'],
           'max_features':['sqrt', 'log2',None]}
grid = GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,n_jobs=-1,verbose=3,scoring='f1_weighted')
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
grid_result = grid.cv_results_
```

```python
table = pd.DataFrame.from_dict(grid_result)
```

```python
table
```

```python
grid.best_params_
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, roc_auc_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

Type1 error is 0 in the above confusion_matrix. And Type2 error is 5. 

```python
clf=classification_report(y_test,y_pred)
print(clf)
```

Accuracy on the above result is 96% which is good. 

```python
f1score = f1_score(y_test,y_pred)
print(f1score)
```

F1score value is closer to 96.8% which is good.

```python
roc_auc_score1 = roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
print(roc_auc_score1)
```
